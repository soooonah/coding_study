```python
#브리또 메뉴 + 옵션 추가
class burritoclass:
    global burrito
    global toping
    global total
    def __init__(self):
        self.name=input('브리또 메뉴를 선택해 주세요> ')
        self.b=burrito.keys()
        self.t=toping.keys()
        self.list=[]
        
    def manuname(self):
        while 1:
            if self.name in self.b:
                self.list.append(self.name)
                break
            else: 
                self.name=input('잘못입력하셨습니다 다시 입력해주세요> ')
                
    def topingset(self):
        mc=input('''           토핑 추가 및 세트메뉴 변경           
1.토핑추가  2.(음료)세트 +500  3.(음료+감튀)세트 +1500  4.변경없음 
> ''')
        topingcount=0
        while 1:
            if mc=='1':
                while 1:
                    topingname=input('추가하실 토핑를 입력하세요(종료시 0) ')
                    if topingname in self.t:
                        self.list.append(topingname) #토핑이름을 리스트에 저장
                        topingcount+=toping[topingname]
                    elif topingname=='0':
                        ms=input('''           세트 선택
    1.(음료)세트 +500  2.(음료+감튀)세트 +1500  3.변경없음
> ''' )
                        while 1:
                            if ms=='1':
                                mc='2'
                                break
                            elif ms=='2':
                                mc='3'
                                break
                            elif ms=='3':
                                mc='4'
                                break
                            else:
                                ms=input('잘못입력하셨습니다 다시 입력해주세요> ')
                        break
                    else:
                        print('잘못입력하셨습니다')
            elif mc=='2':
                self.list.append('(음료)세트')
                self.list.append(burrito[self.name]) # 메뉴가격
                self.list.append(topingcount) #토핑 총액
                self.list.append(500) #세트변경가격
                total.append(self.list)
                break
            elif mc=='3':
                self.list.append('(음료+감튀)세트')
                self.list.append(burrito[self.name]) # 메뉴가격
                self.list.append(topingcount) #토핑 총액
                self.list.append(1500) #세트변경가격
                total.append(self.list)
                break
            elif mc=='4':
                self.list.append('단품')
                self.list.append(burrito[self.name]) # 메뉴가격
                self.list.append(topingcount) #토핑 총액
                self.list.append(0) #세트변경가격
                total.append(self.list)
                break
            else :
                mc=input('잘못입력하셨습니다 다시 입력해주세요> ')
```

```python
#결제
class billclass:
    global total
    global totalfrie
    global pl
    def __init__(self):
        self.phone=input('포인트 전화번호 입력해주세요> (없으면 0) ')
        for i in range(0,len(pl)):
            if self.phone in pl[i][0]:
                p=float(pl[i][1])
                self.point=int(p)              
        self.money=0
    def bill(self):#계산서
        print('---------------------------------------------')    
        print('<                   bill                    >')
        print('---------------------------------------------')
        n=1
        for i in range(0,len(total)): #브리또 주문 보여주기
            name=total[i][0] 
            print('%d.%s'%(n,name)) #번호랑 이름 출력
            print('-'*40,total[i][-3]) #브리또 가격 출력
            self.money+=int(total[i][-3]) 
            if len(total[i][1:-4])!=0: #토핑이 있을경우
                print('+%s 추가 -> %s'%(total[i][1:-4],total[i][-2]))
                self.money+=int(total[i][-2])
            if total[i][-4]=='단품': #단품구매의 경우
                self.money+=int(total[i][-1])
                n+=1
            else: #세트인경우
                print('+%s 추가 -> %s'%(total[i][-4],total[i][-1]))
                self.money+=int(total[i][-1])
                n+=1
            print('---------------------------------------------')
        for i in range(0,len(totalfrie)):
            name=totalfrie[i][0]
            print('%d.%s'%(n,name))
            print('-'*40,totalfrie[i][-1])
            self.money+=totalfrie[i][-1]
        print('---------------------------------------------')
        print('총 결제금액                             %d원'%self.money)
    def poin(self): #포인트 사용 / 적립
        if self.phone != '0':
            point=0
            yp=int(input('''                 포인트  
        1.포인트 적립  2.포인트 사용 
> '''))
            while 1:
                if yp==1:
                    point=self.money*0.03
                    self.point+=point
                    print('%d점 적립되었습니다.'%point)
                    print('%d점 있습니다'%self.point)
                    break
                elif yp==2:
                    if self.point==0:
                        print('포인트가 없습니다')
                        point+=self.money*0.03
                        self.point+=point
                        print('%d점 적립되었습니다.'%point)
                        print('%d점 있습니다'%self.point)
                        break
                    else:
                        print('%d점 사용합니다'%self.point)
                        self.money-=self.point
                        point+=self.money*0.01
                        self.point=point
                        print('%d점 적립되었습니다.'%point)
                        break
                else:
                    print('잘못입력하셨습니다 다시 입력해주세요')
        
        while 1:#결제수단   
            cash=int(input('''          결제 수단 선택           
          1.현금       2.카드 
> '''))
            if cash==1:
                while 1:
                    print('결제금액은 %d원입니다'%self.money)
                    m=int(input('돈을 넣어 주세요 '))
                    m-=self.money
                    if m<0:
                        print('돈이 부족합니다 다시 넣어 주세요')
                    elif m==0:
                        print('거스름돈은 없습니다 맛있게 드세요')
                        break
                    elif m>0:
                        print(m,'원 받아주세요 맛있게 드세요')
                        break
                    else :
                        print('다시 입력해주세요')
            elif cash==2:
                print('카드결제를 완료했습니다. 맛있게 드세요')
                break
            else:
                print('잘못입력하셨습니다 다시 입력해주세요')
        for i in range(0,len(pl)):
            if self.phone in pl[i][0]:
                pl[i][1]=self.point
            pl[i]='%s/%d\n'%(pl[i][0],pl[i][1])
            
        with open('point.txt','w') as p:
            for i in range(0,len(pl)):
                p.write(pl[i])
```

```python
def manu (name,d,a,b):
    n=list(d.keys()) 
    c=list(d.values())
    print('---------------------------------------------')
    print('<                   %s                 >'%name)
    print('---------------------------------------------')
    for i in range(0,len(n)):
        bn=n[i] #메뉴 이름 정렬
        print('%d.%s'%((i+1),bn.ljust(a)),end=' ') 
    print()
    for i in range(0,len(c)):
        bc='%d'%c[i] #메뉴가격 정렬
        print(bc.center(b),end=' ')
    print()
```
